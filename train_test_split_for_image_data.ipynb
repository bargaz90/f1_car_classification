```python
import os
import shutil
import random
```

```python
path = r'D:\_download\Formula1_2019_cars_image_data'
target_path = r'D:\_download\Formula1_2019_cars_image_data\TEST'
test_size = 0.2
cars = os.listdir(path)

for car in cars:
    
    images = os.listdir((os.path.join(path, car)))
    n_images = len(images)
    test_images = random.sample(images, int(n_images * test_size))
    os.mkdir(os.path.join(target_path, car))
    
    for i in range(len(test_images)):
        source_path = os.path.join(path, car, test_images[i])
        destination_path = os.path.join(target_path, car,test_images[i])
        shutil.move(source_path, destination_path)
```
